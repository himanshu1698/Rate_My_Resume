```python
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
def glassdoor_job_search(query, location):
    try:
        jobs = []
        # Set up the Selenium webdriver (make sure you have chromedriver installed and its path added to PATH)
        driver = webdriver.Chrome()

        # Open Glassdoor website
        driver.get("https://www.glassdoor.com/Job/index.htm")

        # try:
            # Find the job search input field and location input field
        job_search_box = driver.find_element("id", "searchBar-jobTitle")
        location_box = driver.find_element("id", "searchBar-location")

        # Enter the search query and location
        job_search_box.clear()
        job_search_box.send_keys(query)
        location_box.clear()
        location_box.send_keys(location)

        # Submit the search form
        location_box.send_keys(Keys.RETURN)

        # Wait for the search results to load
        time.sleep(1)

        # Get job listings
        job_listings = driver.find_elements(By.CLASS_NAME,"JobCard_jobCardContainer__l0svv")

        print(job_listings)
        c = 0
        for job_listing in job_listings:
            # Extract job details (you may need to adjust these depending on the Glassdoor site structure)
            job_title = job_listing.find_element(By.CLASS_NAME,"JobCard_seoLink__WdqHZ").text
            company = job_listing.find_element(By.CLASS_NAME,"EmployerProfile_profileContainer__d5rMb").text
            location = job_listing.find_element(By.CLASS_NAME,"JobCard_location__N_iYE").text
            try:
                salary_estimate = job_listing.find_element(By.CLASS_NAME,"JobCard_salaryEstimate___m9kY").text
            except Exception as e:
                print(str(e))
                salary_estimate = -1
            # date_posted = job_listing.find_element_by_css_selector(".jobFooter>.minor").text

            # Print job details
            print(f"Job Title: {job_title}")
            print(f"Company: {company}")
            print(f"Location: {location}")
            print("Salary Estimate: {}".format(salary_estimate))
            # print(f"Date Posted: {date_posted}")
            c += 1
            print(c)
            jobs.append([job_title,company,location,salary_estimate])
            print("\n")



        show_more = driver.find_element(By.CLASS_NAME,'button_Button__meEg5') 
        print(show_more)
        show_more.click()    
        time.sleep(1)


        if driver.find_element(By.CLASS_NAME,'ModalContainer'):
            close = driver.find_element(By.CLASS_NAME,'CloseButton')
            close.click()
            
        return pd.DataFrame(jobs, columns=['Job_Title','Company','Location','Salary_Estimate'])

    

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        # pass
        # Close the browser window
        driver.quit()
```

```python
data = glassdoor_job_search("Data Scientist", "")
```
